```python
class Solution:
    def isPalindrome(self, x: int) -> bool:
        X = x
        rev = 0

        while (x>0):
            temp = x % 10
            rev = rev *10 + temp
            x = x//10
        return rev == X

if __name__ == "__main__":
    %timeit test = Solution().isPalindrome(12321)
    print(test)
```

Best Time to Buy and Sell Stock

```python
class Solution:
    def maxProfit(self, prices: list[int]) -> int:

        m_price = prices[0]
        M_profit = 0
        
        for p in prices:
            if p < m_price:
                m_price = p
            elif p - m_price > M_profit:
                M_profit = p - m_price
        return M_profit

    

                
        

if __name__ == "__main__":
    test = Solution().maxProfit([7,1,5,3,6,4,9,5,43,6,4,3,5,6,4,33,222])
    print(test)
```

Valid Parentheses

```python
class Solution(object):
    def isValid(self, s: str) -> bool:
        d = {'(':')', '{':'}','[':']'}
        stack = []
        for i in s:
            if i in d:
                stack.append(i)
            elif len(stack) == 0 or d[stack.pop()] != i:
                return False
        return len(stack) == 0

if __name__ == "__main__":
    test = Solution().isValid("{}(())[]")
    print(test)
```

Maximum Subarray

```python
class Solution(object):
    def maxSubArray(self, nums):
        for i in range(0, len(nums)-1):
            if nums[i] > 0:
                nums[i+1] += nums[i]
        return max(nums)

if __name__ == "__main__":
    test = Solution().maxSubArray([-2,1,-3,4,-1,2,1,-5,4,8])
    print(test)
            
```

Merge Intervals

```python
class Solution:
    def merge(self, intervals: list[list[int]]) -> list[list[int]]:
        minHeap = intervals
        minHeap.sort()  # Heapify in increasing by the starting time order
        minHeap = minHeap[::-1]
        ans = []
        while minHeap:
            start, end = minHeap.pop()
            if minHeap:
                while end >= minHeap[len(minHeap)-1][0]:
                    end = max(minHeap[len(minHeap)-1][1],end)
                    minHeap.pop()
                    if not minHeap:
                        break
                    
            ans.append([start,end])
        return ans

if __name__ == "__main__":
    test = Solution().merge([[1,4],[4,5]])
    print(test)
```

```python
class Algorithms:
    
    def __init__(self):
        pass

    # 3.A. binary search through iteration
    def binarySearchIterator(self, arr, find):
        
        low = 0
        high = len(arr)

        while(high > low):
            mid = (low+high)//2
            if find < arr[mid]:
                high = mid-1
            elif find > arr[mid]:
                low = mid +1
            else:
                return mid

        return -1
    
    # 4.B. Merge sort:
    def merge_sort(self, my_list):
        if len(my_list) > 1:
            mid = len(my_list) // 2
            left = my_list[:mid]
            right = my_list[mid:]

            # Recursive call on each half
            self.merge_sort(left)
            self.merge_sort(right)

            # Two iterators for traversing the two halves
            i = 0
            j = 0
            
            # Iterator for the main list
            k = 0
            
            while i < len(left) and j < len(right):
                if left[i] < right[j]:
                # The value from the left half has been used
                    my_list[k] = left[i]
                    # Move the iterator forward
                    i += 1
                else:
                    my_list[k] = right[j]
                    j += 1
                # Move to the next slot
                k += 1

            # For all the remaining values
            while i < len(left):
                my_list[k] = left[i]
                i += 1
                k += 1

            while j < len(right):
                my_list[k]=right[j]
                j += 1
                k += 1
        return my_list


    def find_minimum(self, lst):
        if (len(lst) <= 0):
            return None
        return self.merge_sort(lst)  # sort list
        return lst[0]  # return first element
             

if __name__ == "__main__":
    test = Algorithms()
    print(f"Merge Sort: LIST: {test.find_minimum([2,34,5,2,54,2,4,9,0,5,23,4,765,354,4])}" )
    print(f"Binary Search Iterator: INDEX: {test.binarySearchIterator([1,3,4,6,9,12,34,56,78,123,543], 4)}" )
    
    
```
